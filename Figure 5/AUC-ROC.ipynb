```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
import csv
import sys
import numpy as np

def ro_curve(y_pred, y_label, figure_file, method_name):
    y_label = np.array(y_label)
    y_pred = np.array(y_pred)    
    fpr = dict()
    tpr = dict() 
    roc_auc = dict()
    fpr[0], tpr[0], thresholds = roc_curve(y_label, y_pred)
    roc_auc[0] = auc(fpr[0], tpr[0])
    lw = 1
    plt.plot(fpr[0], tpr[0],
         lw=1, label= method_name + ' (AUC = %0.4f)' % roc_auc[0])
    plt.plot([0, 1], [0, 1], color='black', lw=1, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    fontsize = 14
    plt.xlabel('False Positive Rate', fontsize = 14, fontfamily='Arial')
    plt.ylabel('True Positive Rate', fontsize = 14, fontfamily='Arial')
    plt.legend(loc="lower right", fontsize='12')
    plt.rcParams['figure.figsize'] = (8, 6)
    plt.savefig(figure_file + "_new2"+".tiff",dpi=600)
    plt
    return 

def col_pic():
    for i in range(0,2):
        y_label = []
        y_pred = []
        k=['molecular docking', 'ensemble model']
        with open(str(k[i]) + ".csv") as f:
            f1 = csv.reader(f)
            for line in f1:
                y_label.append(int(float(line[0])))
                y_pred.append(float(line[1]))
            ro_curve(y_pred,y_label,"auc_val",k[i])
        

def main():
    col_pic()
    
if __name__=="__main__":
    main() 
```
